```python
data = {
    "Salary" : [25000, 30000, 37000, 28000, 39000, 48000, 55000, 52000, 35000, 26000, 27000, 32000, 33000, 28000, 40000, 36000, 34000, 45000, 47000, 50000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
# Line Plot:
plt.plot(df["Salary"], color = "red", marker = "o", linestyle = ":", linewidth = "2")
plt.grid()
plt.show()
```

```python
# Histogram
plt.hist(df['Salary'], color= 'orange')
plt.show()
```

```python
# Boxplot:

plt.boxplot(df["Salary"])
plt.show()
```

```python
df.loc[20] = [0] #outlier
```

```python
plt.boxplot(df["Salary"])
plt.show()
```

```python
df.drop(index = 20, inplace=True)
```

```python
# Univariate : Categorical
```

```python
df["dept"] = ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR'] * 2
```

```python
df.head()
```

```python
df.info()
```

```python
# Pie Chat:
count = df['dept'].value_counts()
count
```

```python
plt.pie(count, labels = count.index, autopct= '%1.1f', explode= [0, 0.1, 0.2])
plt.axis('equal') # pie chat in complete circle
plt.show()
```

```python
# Countplot

plt.bar(count.index, count, color = ['cyan', 'orange', 'yellow'])
plt.show()
```

```python
# Bivariate - numerical - numerical
```

```python
df.head()
```

```python
df["Age"] = [22, 25, 29, 24, 30, 35, 40, 36, 28, 22, 23, 25, 26, 24, 31, 29, 27, 33, 35, 38]
```

```python
df.head()
```

```python
# Scatter Plot:

plt.scatter(df["Age"], df["Salary"], color = 'orange')
plt.show()
```

this is also know as Positive Corelation

```python
sort_age = df.sort_values("Age")
```

```python
# Line Plot:
plt.plot(sort_age["Age"], df["Salary"], color = 'red', marker = "*",  linewidth = '2')
plt.grid()
plt.show()
```

```python
# Bar Chart:

plt.bar(sort_age["Age"], df["Salary"], color = 'green', align="center")
plt.show()
```

```python
# Bivariate: Numerical - categorical
```

```python
hr_sal = df[df["dept"] == "HR"]["Salary"]
```

```python
hr_sal
```

```python
it_sal = df[df["dept"] == "IT"]["Salary"]
it_sal
```

```python
finance_sal = df[df["dept"] == "Finance"]["Salary"]
finance_sal
```

```python
# Boxplot:

plt.boxplot([hr_sal, it_sal, finance_sal], labels= ['HR', 'IT', 'Finance'])
plt.show()
```

orange line represents median salary

```python
# Pie chart:

salary_by_dept = df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct= '%1.2f', shadow= True, explode = [0.1, 0, 0.1])
plt.axis('equal') # pie chat in complete circle
plt.show()
```

```python
# bar plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
finance_mean = sum(finance_sal)/len(finance_sal)
```

```python
plt.bar(['HR', 'IT', 'Finance'], [hr_mean, it_mean, finance_mean], color = ['blue', 'green', 'red'])
plt.grid()
plt.show()
```

```python
# Multivariate Analysis: 3 numerical columns
```

```python
df.head()
```

```python
df["experience"] = [1, 2, 4, 1.5, 5, 8, 12, 10, 4, 1, 1.2, 2, 3, 1.8, 5.5, 4.5, 3.2, 7, 8, 10]
```

```python
df.head()
```

```python
# Bubble Plot:

plt.scatter(df["Age"], df["Salary"], s = df["experience"]*30, c = "skyblue", edgecolor = "red")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# 2 numerical and categorical column
```

```python
plt.scatter(df["Age"], df["Salary"], c = df["dept"].map({"HR": "yellow", "IT": "blue", "Finance": "orange"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Dept")
plt.legend()
plt.show()
```

```python
color = {"HR" : "yellow", "IT": "blue", "Finance" : "orange"}

for dept, color in color.items():
  df_dept = df[df["dept"] == dept]
  plt.scatter(df_dept["Age"], df_dept["Salary"], c = color, label = dept)

  plt.legend()
```

```python
# Object Oriented API:
# multiple plots or sub plots in a single figure
```

```python
fig, axs = plt.subplots(1, 3, figsize = (18, 5))

# 1. Line plot

axs[0].plot(sort_age["Age"], df["Salary"], color = "red", marker = "*", linewidth = "2", markersize = 2)
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# 2. Histogram:

axs[1].hist(df["Salary"], bins = 5, color = "skyblue")
axs[1].grid()
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")


# 3. Boxplot:

axs[2].boxplot(df["Salary"])
axs[2].grid()
axs[2].set_title("Boxplot")
axs[2].set_xlabel("Salary")
axs[2].set_ylabel("Salary")


# In-built Feature in matplotlib to save a figure:
plt.savefig("multiplots.png")


plt.show()
```

```python
data = {
    'Year': [2020, 2021, 2022, 2023],
    'Sales': [1000, 1200, 1500, 1800],
    'Profit': [500, 600, 700, 800],
    'Expenses': [200, 300, 400, 500]
}
```
